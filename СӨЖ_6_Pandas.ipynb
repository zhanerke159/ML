Барлық код жазатын ұяшықтарда `# Кодты осында жазыңыз` деген пікір (комментарий) қалдырылған. Сол жерге өз шешіміңізді жазып, ұяшықты іске қосыңыз.

## 1-бөлім. Деректерді дайындау (DataFrame құру)

**Бөлімнің мақсаты:** Python сөздігі (dictionary) негізінде екі өлшемді кесте (DataFrame) жасап үйрену және кестенің бастапқы құрылымымен танысу.

Төменде 5 бағаннан және 5 қатардан тұратын сөздік берілген. Бұл деректер студенттердің Машиналық оқыту (ML) және Математика пәндерінен алған бағаларын көрсетеді.

```
import pandas as pd
import numpy as np

data = {
    'Студент': ['Асан', 'Берік', 'Сәуле', 'Данияр', 'Еңлік'],
    'Жасы': [20, 21, 19, 22, 20],
    'Қаласы': ['Алматы', 'Астана', 'Шымкент', 'Алматы', 'Астана'],
    'ML_ұпайы': [85, 92, 78, 95, 88],
    'Math_ұпайы': [90, 88, 85, 92, 95]
}
```

**Тапсырма 1:** Жоғарыдағы `data` сөздігінен `df` атты DataFrame объектісін жасаңыз.

```
# Кодты осында жазыңыз
df = pd.DataFrame(data)
```

**Тапсырма 2:** Жасалған `df` кестесінің алғашқы 3 жолын ғана экранға шығарыңыз.

```
# Кодты осында жазыңыз
print(df.head(3))
```

## 2-бөлім. Series және DataFrame, Индексация және таңдау

**Бөлімнің мақсаты:** Кестеден белгілі бір бағандарды бөліп алу, қатарлар бойынша нақты деректерді индекстері арқылы (`loc`, `iloc`) дұрыс таңдап алу дағдыларын қалыптастыру.

**Тапсырма 3:** Кестеден тек 'Студент' және 'ML_ұпайы' бағандарын таңдап, оны жаңа айнымалыға сақтап, экранға шығарыңыз.

```
# Кодты осында жазыңыз
new = df[['Студент', 'ML_ұпайы']]

print(new)
```

**Тапсырма 4:** `iloc` әдісін қолданып, кестенің 2-ші және 3-ші жолдарын (индекстері 1 және 2 болатын жолдар) шығарыңыз.

```
# Кодты осында жазыңыз
print(df.iloc[1:3])
```

**Тапсырма 5:** `loc` әдісін қолданып, 'Сәуле' (қатар индексі 2) есімді студенттің барлық мәліметтерін шығарыңыз.

```
# Кодты осында жазыңыз
print(df.loc[2])
```

**Тапсырма 6:** 'Қаласы' бағанын таңдап, оның типі Pandas Series екеніне көз жеткізу үшін `type()` функциясын қолданыңыз.

```
# Кодты осында жазыңыз
c = df['Қаласы']
print(type(c))
```

**Тапсырма 7:** Кестенің өлшемін (жолдар мен бағандар санын) және жалпы техникалық ақпаратын экранға шығарыңыз.

```
# Кодты осында жазыңыз
print(df.shape)
print(df.info())
```

## 3-бөлім. Шарт бойынша сүзу (Filtering)

**Бөлімнің мақсаты:** Логикалық шарттарды (үлкен, кіші, тең, және/немесе) қолдана отырып, үлкен деректер базасынан тек өзімізге қажетті ақпаратты сүзіп алуды үйрену.

**Тапсырма 8:** Жасы 20-дан үлкен студенттердің тізімін шығарыңыз.

```
# Кодты осында жазыңыз
print(df[df['Жасы']>20])
```

**Тапсырма 9:** ML ұпайы 90-нан жоғары **немесе** Math ұпайы 90-нан жоғары студенттерді табыңыз (логикалық `|` операторын қолданыңыз).

```
# Кодты осында жазыңыз
print(df[df['ML_ұпайы']>90 | (df['Math_ұпайы'] > 90)])
```

**Тапсырма 10:** Тек 'Алматы' қаласынан келген студенттерді көрсетіңіз.

```
# Кодты осында жазыңыз
print(df[df['Қаласы']=='Алматы'])
```

**Тапсырма 11:** 'Астана' қаласынан келген **және** ML ұпайы 85-тен жоғары студенттерді табыңыз (логикалық `&` операторын қолданыңыз).

```
# Кодты осында жазыңыз
print(df[(df['Қаласы']>'Астана') & (df['ML_ұпайы']>85)])
```

**Тапсырма 12:** `isin()` әдісін қолданып, 'Шымкент' немесе 'Алматы' қалаларында тұратын студенттерді сүзіп алыңыз.

```
# Кодты осында жазыңыз
print(df[df['Қаласы'].isin(['Шымкент', 'Алматы'])])
```

## 4-бөлім. Топтастыру (groupby) және агрегаттау

**Бөлімнің мақсаты:** Деректерді белгілі бір санаттар бойынша топтастыру және оларға математикалық/статистикалық (қосынды, орташа мән, максимум) функцияларды қолдану.

**Тапсырма 13:** Әр қалада қанша студент бар екенін санаңыз (`groupby` және `size` немесе `count` қолданыңыз).

```
# Кодты осында жазыңыз
print(df.groupby('Қаласы').size())
```

**Тапсырма 14:** Қалалар бойынша топтастырып, ML ұпайының орташа мәнін (`mean`) табыңыз.

```
# Кодты осында жазыңыз
print(df.groupby('Қаласы')['ML_ұпайы'].mean())
```

**Тапсырма 15:** Жасы бойынша топтастырып, студенттердің Математикадан алған ең жоғарғы (`max`) ұпайларын анықтаңыз.

```
# Кодты осында жазыңыз
print(df.groupby('Жасы')['Math_ұпайы'].max())
```

**Тапсырма 16:** `agg()` функциясын қолданып, бүкіл кестедегі ML ұпайының әрі орташа, әрі максималды мәнін бір уақытта есептеңіз.

```
# Кодты осында жазыңыз
print(df.groupby('Қаласы').agg({'ML_ұпайы': ['mean', 'max']}))
```

**Тапсырма 17:** Қалалар бойынша топтастырып, барлық сандық бағандардың (Жасы, ML_ұпайы, Math_ұпайы) қосындысын (`sum`) шығарыңыз.

```
# Кодты осында жазыңыз
print(df.groupby('Қаласы')[['Жасы', 'ML_ұпайы', 'Math_ұпайы']].sum())
```

## 5-бөлім. Кестелерді біріктіру (merge, join, concat)

**Бөлімнің мақсаты:** Бірнеше бөлек кестелерде (DataFrames) сақталған мәліметтерді ортақ кілттер немесе индекстер арқылы бір кестеге біріктіру әдістерін меңгеру.

Бұл бөлім үшін бізге қосымша кесте қажет. Төмендегі кодты орындап, жаңа кестені жасаңыз:

```
# Бұл ұяшықты жай ғана іске қосыңыз
data_stipend = {
    'Студент': ['Асан', 'Берік', 'Марат'],
    'Стипендия': [41000, 46000, 41000]
}
```

**Тапсырма 18:** `data_stipend` сөздігінен `df_stipend` атты екінші DataFrame жасаңыз.

```
# Кодты осында жазыңыз
df_stipend = pd.DataFrame(data_stipend)

print(df_stipend)
```

**Тапсырма 19:** `df` және `df_stipend` кестелерін 'Студент' бағаны бойынша іштей біріктіріңіз (`pd.merge`, `how='inner'`).

```
# Кодты осында жазыңыз
print(pd.merge(df, df_stipend, on='Студент', how='inner'))
```

**Тапсырма 20:** Осы екі кестені сол жақ біріктіру (`how='left'`) арқылы қосып, нәтижесін `df_left` айнымалысына сақтаңыз және шығарыңыз. (Стипендиясы жоқ студенттерде `NaN` пайда болатынына назар аударыңыз).

```
# Кодты осында жазыңыз
df_left = pd.merge(df, df_stipend, on='Студент', how='left')

print(df_left)
```

**Тапсырма 21:** Жаңа бір студенттің мәліметтері бар кішігірім DataFrame жасап (`df_new`), оны негізгі `df` кестесінің астына жалғаңыз (`pd.concat`).

```
# Кодты осында жазыңыз
df_new = pd.DataFrame({
    'Студент': ['Марат'],
    'Жасы': [21],
    'Қаласы': ['Қарағанды'],
    'ML_ұпайы': [84],
    'Math_ұпайы': [89]
})

df_concat = pd.concat([df, df_new], ignore_index=True)

print(df_concat)
```

**Тапсырма 22:** Индекстер арқылы біріктіру үшін `join()` әдісін қолданып көріңіз. (Ол үшін алдымен екі кестенің де индексін `set_index('Студент')` арқылы өзгертіп алу қажет).

```
# Кодты осында жазыңыз
df_indexed = df.set_index('Студент')
df_stipend_indexed = df_stipend.set_index('Студент')

print(df_indexed.join(df_stipend_indexed))
```

## 6-бөлім. Қорытынды кешенді тапсырмалар

**Бөлімнің мақсаты:** Алдыңғы бөлімдерде алған білімді біріктіріп, нақты аналитикалық шешімдер қабылдауға арналған күрделірек тапсырмаларды орындау.

**Тапсырма 23:** Студенттердің ML және Математика ұпайларының қосындысынан тұратын жаңа 'Жалпы_ұпай' бағанын негізгі `df` кестесіне қосыңыз.

```
# Кодты осында жазыңыз
df['Жалпы_ұпай'] = df['ML_ұпайы'] + df['Math_ұпайы']

print(df)
```

**Тапсырма 24:** Кестені жаңадан қосылған 'Жалпы_ұпай' бағаны бойынша кему ретімен (үлкеннен кішіге қарай) сұрыптаңыз (`sort_values`).

```
# Кодты осында жазыңыз
df_sorted = df.sort_values(by='Жалпы_ұпай', ascending=False)

print(df_sorted)
```

**Тапсырма 25:** Жалпы ұпайы ең жоғары болған үздік студенттің **тек қана** аты мен қаласын экранға шығаратын код жазыңыз.

```
# Кодты осында жазыңыз
top_student = df.sort_values(by='Жалпы_ұпай', ascending=False).iloc[0][['Студент', 'Қаласы']]

print(top_student)
```
