**Мақсаты:** Python тіліндегі негізгі деректер құрылымдарымен жұмыс істеу дағдыларын бекіту және тізімдік өрнектер арқылы кодты оңтайландыруды үйрену.




## Қысқаша теория

### 1. Тізімдер (Lists)
Тізім  реттелген, өзгермелі элементтер жиынтығы.

* Синтаксисі: `my_list = [1, 2, 3]`
* Ерекшелігі: Элементтерді қосуға, өшіруге және өзгертуге болады.

### 2. Кортеждер (Tuples)
Кортеж реттелген, бірақ **өзгермейтін** (immutable) элементтер жиынтығы.

* Синтаксисі: `my_tuple = (1, 2, 3)`
* Ерекшелігі: Құрылғаннан кейін элементтерін өзгерту мүмкін емес.

### 3. Жиындар (Sets)
Жиын реттелмеген және **қайталанбайтын** элементтер жиынтығы.
* Синтаксисі: `my_set = {1, 2, 3}`
* Ерекшелігі: Математикалық жиын операцияларын (біріктіру, қиылысу) орындауға ыңғайлы.

### 4. Сөздіктер (Dictionaries)
Сөздік  "кілт-мән" (key-value) жұптарынан тұратын реттелмеген жиынтық.
* Синтаксисі: `my_dict = {"name": "Ali", "age": 20}`
* Ерекшелігі: Мәліметті кілт арқылы жылдам іздеуге арналған.

### 5. Тізімдік өрнектер (List Comprehension)
Бұл тізімдерді құрудың қысқа әрі тиімді жолы. `for` циклі мен `if` шарттарын бір жолға сыйғызуға мүмкіндік береді.

* Мысалы: `[x**2 for x in range(10) if x % 2 == 0]`


**9-тапсырма: Сөздіктер және циклдар**
Төмендегі сөздікте тауарлар мен олардың бағасы берілген. Бағасы 500-ден жоғары тауарлардың атын ғана шығарыңыз.
`products = {"nan": 150, "sut": 600, "may": 800, "tuz": 80}`


```
products = {"nan": 150, "sut": 600, "may": 800, "tuz": 80}

for name, price in products.items():
    if price > 500:
        print(name)
```


**10-тапсырма: Мәтінді талдау**
Берілген сөйлемдегі сөздердің ұзындығын есептейтін сөздік құрыңыз (кілт - сөз, мән - ұзындық).
Сөйлем: `"Python өте қызықты тіл"`

```
sentence = "Python өте қызықты тіл"
words = sentence.split()
lengths = {w: len(w) for w in words}
print(lengths)
```


**11-тапсырма: Жиындар операциясы**
Екі топ студенттерінің тізімі берілген.
`group_a = {"Ali", "Bob", "Dan"}`
`group_b = {"Dan", "Eva", "Ali"}`
Екі топта да бар ортақ студенттерді табыңыз (қиылысу - intersection).


```
group_a = {"Ali", "Bob", "Dan"}
group_b = {"Dan", "Eva", "Ali"}
common = group_a & group_b
print(common)
```


**13-тапсырма: Тізім ішіндегі тізімдер (Matrix)**
3x3 өлшемді матрицаны (тізім ішіндегі тізім) List Comprehension арқылы жасаңыз. Элементтері 0-ден 8-ге дейін ретпен орналассын.
Күтілетін нәтиже: `[[0, 1, 2], [3, 4, 5], [6, 7, 8]]`


```
matrix = [[i*3 + j for j in range(3)] for i in range(3)]
print(matrix)
```


**14-тапсырма: Жолдарды өңдеу**

`words = ["hi", "hello", "is", "python", "go"]`


```
words = ["hi", "hello", "is", "python", "go"]
result = [w.upper() for w in words if len(w) > 3]
print(result)
```


### III. Қиын деңгей

**15-тапсырма: Сөздердің жиілігін есептеу**
Төмендегі мәтінде әр сөздің қанша рет кездесетінін есептейтін бағдарлама жазыңыз (нүкте, үтірлерді ескермеңіз). `get()` әдісін қолданыңыз.
Мәтін: `"data science is science and data is data"`


```
text = "data science is science and data is data"
freq = {}

for w in text.split():
    freq[w] = freq.get(w, 0) + 1

print(freq)
```


**16-тапсырма: Күрделі сұрыптау**
Студенттер тізімі берілген (аты және ұпайы бар кортеждер). Студенттерді ұпайы бойынша кему ретімен сұрыптаңыз.
`students = [("Ali", 85), ("Bota", 92), ("Daulet", 78)]`
`lambda` функциясын қолданыңыз.


```
students = [("Ali", 85), ("Bota", 92), ("Daulet", 78)]
sorted_students = sorted(students, key=lambda x: x[1], reverse=True)
print(sorted_students)
```


**17-тапсырма: Nested List Comprehension**
Төмендегідей құрылымды тізімдік өрнек арқылы жасаңыз ("flattening" - жазық тізімге айналдыру):
Кіріс дерек: `[[1, 2, 3], [4, 5, 6], [7, 8, 9]]`
Шығыс дерек: `[1, 2, 3, 4, 5, 6, 7, 8, 9]`


```
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [x for row in data for x in row]
print(flat)
```


**18-тапсырма: Сөздіктерді біріктіру**
Екі сөздік берілген:
`d1 = {'a': 100, 'b': 200, 'c':300}`
`d2 = {'a': 300, 'b': 200, 'd':400}`
Бұл екі сөздікті біріктіріңіз. Егер кілттер сәйкес келсе, олардың мәндерін қосыңыз.


```
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}

merged = d1.copy()
for k, v in d2.items():
    merged[k] = merged.get(k, 0) + v

print(merged)
```


**19-тапсырма: Жай сандарды табу (List Comprehension)**
2-ден 50-ге дейінгі сандардың ішінен жай сандарды (тек 1-ге және өзіне бөлінетін) табатын бір жолдық list comprehension жазыңыз. `all()` функциясын қолдануға болады.


```
primes = [n for n in range(2, 51) if all(n % d != 0 for d in range(2, int(n**0.5) + 1))]
print(primes)
```


**20-тапсырма: Деректерді сүзу және түрлендіру**
Сізде қызметкерлер тізімі бар:
`employees = [{'name': 'A', 'salary': 1000}, {'name': 'B', 'salary': 500}, {'name': 'C', 'salary': 1200}]`
Жалақысы 800-ден жоғары қызметкерлердің атын тізімге жинаңыз, бірақ аттарын "Mr./Ms. " префиксімен жазыңыз.


```
employees = [{'name': 'A', 'salary': 1000}, {'name': 'B', 'salary': 500}, {'name': 'C', 'salary': 1200}]

result = [f"Mr./Ms. {e['name']}" for e in employees if e['salary'] > 800]
print(result)
```
