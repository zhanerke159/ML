## Тапсырма 6: ML Алдын Ала Өңдеу Функциялары

**Мақсат:** Деректерді нормализациялайтын және стандарттайтын функциялар жазу.

Нормализация мен стандарттау — ML алгоритмдерінің дұрыс жұмыс істеуі үшін маңызды алдын ала өңдеу қадамдары. Min-Max нормализациясы мәндерді [0, 1] диапазонына, Z-score стандарттауы орташа мәнді 0, стандартты ауытқуды 1 етіп түрлендіреді. Бұл SVM, KNN, нейрондық желі сияқты алгоритмдер үшін өте маңызды.

**Шарт:**

Екі функция жаз:

**`min_max_normalize(arr)`** — Min-Max нормализациясы:
$$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$$

**`z_score_standardize(arr)`** — Z-score стандарттауы:
$$x_{std} = \frac{x - \mu}{\sigma}$$

Екі функцияны да `[2, 4, 4, 4, 5, 5, 7, 9]` массивіне қолдан және нәтижені шығар.

````python
# Тапсырма 6: Нормализация және стандарттау функциялары
import numpy as np
# КОДТЫ ОСЫНДА ЖАЗ
arr = np.array([2, 4, 4, 4, 5, 5, 7, 9])
def min_max_normalize(arr):
  arr = np.array(arr, dtype=float)
  minn = np.min(arr)
  maxx = np.max(arr)
  return (arr-minn)/(maxx - minn)
def z_score_standardize(arr):
  arr = np.array(arr, dtype=float)
  mean = np.array(arr)
  std = np.std(arr)
  return (arr-mean)/std
print(min_max_normalize(arr))
print(z_score_standardize(arr))


````

## Тапсырма 7: Рекурсия және Lambda Функциялар

**Мақсат:** Рекурсивті функция және lambda өрнектерін ML контекстінде қолдану.

Lambda функциялар — атаусыз шағын функциялар, ML-де деректерді трансформациялау үшін `map()`, `filter()`, `sorted()` функцияларымен жиі қолданылады. Рекурсия болса күрделі алгоритмдерді (мысалы, шешім ағашының тереңдігін есептеу) элегантты жазуға мүмкіндік береді.

**Шарт:**

1. **Рекурсивті функция:** `factorial(n)` — n! санын рекурсия арқылы есепте.
2. **Lambda:** `[0.1, 0.5, 0.9, 1.5, -0.3, 2.0]` тізімінен тек `[0, 1]` диапазонындағы мәндерді `filter` + `lambda` арқылы алып шығар.
3. **Lambda + map:** `[-2, -1, 0, 1, 2]` тізіміндегі әр элементке ReLU активация функциясын `map` + `lambda` арқылы қолдан.

> ReLU(x) = max(0, x)

````python
# Тапсырма 7: Рекурсия және Lambda функциялар

# КОДТЫ ОСЫНДА ЖАЗ
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
n = [0.1, 0.5, 0.9, 1.5, -0.3, 2.0]
f = list(filter(lambda x: 0 <= x <= 1, n))
print(f)

v = [-2, -1, 0, 1, 2]
r = list(map(lambda x: max(0, x), v))
print(r)
````

## Тапсырма 9: Мұрагерлік (Inheritance) — ML Модель Иерархиясы

**Мақсат:** Python-да кластар иерархиясын (мұрагерлік) пайдаланып ML модельдерін модельдеу.

Мұрагерлік (inheritance) — OOP-тың негізгі принципі. scikit-learn барлық модельдерін `BaseEstimator` базалық класынан туындатады, бұл `fit()`, `predict()`, `score()` сияқты ортақ интерфейсті қамтамасыз етеді. Осы паттернді түсіну ML кітапханаларының архитектурасын жобалауға негіз береді.

**Шарт:**

1. **`BaseModel`** — базалық класс:
   - `__init__(self, model_name)` — модель атын сақтайды
   - `fit(X, y)` — `NotImplementedError` шығарады (абстрактілі)
   - `predict(X)` — `NotImplementedError` шығарады (абстрактілі)
   - `describe()` — модель туралы ақпаратты шығарады

2. **`LinearRegressionModel(BaseModel)`** — туынды класс:
   - `fit(X, y)` — `np.linalg.lstsq` арқылы коэффициенттерді табады
   - `predict(X)` — `X @ weights` арқылы болжайды
   - `describe()` — базалық сипаттаманы + коэффициенттерді шығарады

3. Тест деректерінде `fit` → `predict` → `describe` тізбегін шақыр.

````python
# Тапсырма 9: Мұрагерлік — ML Модель Иерархиясы

# Тест деректері


# КОДТЫ ОСЫНДА ЖАЗ

import numpy as np

class BaseModel:
    def __init__(self):
        pass
    def fit(self, X, y):
        raise NotImplementedError
    def predict(self, X):
        raise NotImplementedError

class RegressionModel(BaseModel):
    def __init__(self):
        super().__init__()
        self.coef_ = None
        self.intercept_ = None
    def fit(self, X, y):
        X_b = np.column_stack([np.ones(X.shape[0]), X])
        theta = np.linalg.pinv(X_b.T @ X_b) @ X_b.T @ y
        self.intercept_ = theta[0]
        self.coef_ = theta[1:]
    def predict(self, X):
        return X @ self.coef_ + self.intercept_

class ClassificationMetrics:
    def __init__(self, y_true, y_pred):
        self.y_true = y_true
        self.y_pred = y_pred
    def accuracy(self):
        return sum(t==p for t,p in zip(self.y_true,self.y_pred))/len(self.y_true)
    def precision(self):
        TP = sum((t==1 and p==1) for t,p in zip(self.y_true,self.y_pred))
        FP = sum((t==0 and p==1) for t,p in zip(self.y_true,self.y_pred))
        return TP/(TP+FP) if (TP+FP)>0 else 0
    def recall(self):
        TP = sum((t==1 and p==1) for t,p in zip(self.y_true,self.y_pred))
        FN = sum((t==1 and p==0) for t,p in zip(self.y_true,self.y_pred))
        return TP/(TP+FN) if (TP+FN)>0 else 0
    def f1_score(self):
        p = self.precision()
        r = self.recall()
        return 2*(p*r)/(p+r) if (p+r)>0 else 0
    def report(self):
        print(f"Accuracy : {self.accuracy():.3f}")
        print(f"Precision: {self.precision():.3f}")
        print(f"Recall   : {self.recall():.3f}")
        print(f"F1-score : {self.f1_score():.3f}")

x_train = np.column_stack([np.ones(20), np.random.randn(20)])
y_train = 3 * x_train[:,1] + 5 + np.random.randn(20)*0.5
reg = RegressionModel()
reg.fit(x_train[:,1].reshape(-1,1), y_train)
y_pred = reg.predict(x_train[:,1].reshape(-1,1))
y_true = (y_train > np.median(y_train)).astype(int)
y_pred_class = (y_pred > np.median(y_pred)).astype(int)
metrics = ClassificationMetrics(y_true, y_pred_class)
metrics.report()
x_train = np.column_stack([np.ones(20), np.random.randn(20)])  # Бір фичалы деректер
y_train = 3 * x_train[:, 1] + 5 + np.random.randn(20) * 0.5  # y = 3x + 5 + шу
metrics = ClassificationMetrics(y_true, y_pred)
metrics.report()

````

##Тапсырма 10: Толыққанды ML Pipeline Класы

**Мақсат:** Нақты ML жүйесін модельдейтін толыққанды Pipeline класын жазу.

ML Pipeline  деректерді алдын ала өңдеу, модельді оқыту және бағалау қадамдарын автоматты реттілікпен орындайтын құрылым. sklearn-дің `Pipeline` класы осындай принципте жұмыс істейді. Мұны өз бетімен жүзеге асыру бүкіл ML жұмыс ағынын терең түсінуге мүмкіндік береді.

**Шарт:**

`MLPipeline` класы жаз, ол мынадай кезең-кезеңді іске асырсын:

1. **`load_data(X, y)`** — деректерді сақтайды
2. **`train_test_split(test_ratio=0.2)`** — деректерді оқыту/тест жиынтықтарына бөледі
3. **`normalize()`** — оқыту жиынтығы бойынша Min-Max нормализациясын жасайды (тест жиынына да қолданады)
4. **`train()`** — `np.linalg.lstsq` арқылы линейлі регрессия оқытады
5. **`evaluate()`** — MSE және R² метрикаларын есептейді
6. **`run()`** — барлық қадамдарды дұрыс ретпен шақырады

Тест деректерінде толық `run()` жүргіз және нәтижені шығар.

> **Кеңес:** R² = 1 - SS_res / SS_tot, мұндағы SS_res = Σ(y - ŷ)², SS_tot = Σ(y - ȳ)²

````python
# Тапсырма 10: Толыққанды ML Pipeline класы

# Тест деректері
np.random.seed(42)
n_samples = 100
X_data = np.random.randn(n_samples, 3)  # 100 үлгі, 3 белгі
y_data = 2 * X_data[:, 0] - 1.5 * X_data[:, 1] + 0.5 * X_data[:, 2] + np.random.randn(n_samples) * 0.3

# КОДТЫ ОСЫНДА ЖАЗ
import numpy as np

class MLPipeline:
    def __init__(self):
        self.X = None
        self.y = None
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        self.X_min = None
        self.X_max = None
        self.coef_ = None
        self.intercept_ = None

    def load_data(self, X, y):
        self.X = np.array(X, dtype=float)
        self.y = np.array(y, dtype=float)

    def train_test_split(self, test_ratio=0.2):
        n = self.X.shape[0]
        idx = np.arange(n)
        np.random.shuffle(idx)
        split = int(n*(1-test_ratio))
        train_idx, test_idx = idx[:split], idx[split:]
        self.X_train = self.X[train_idx]
        self.X_test = self.X[test_idx]
        self.y_train = self.y[train_idx]
        self.y_test = self.y[test_idx]

    def normalize(self):
        self.X_min = self.X_train.min(axis=0)
        self.X_max = self.X_train.max(axis=0)
        self.X_train = (self.X_train - self.X_min) / (self.X_max - self.X_min)
        self.X_test = (self.X_test - self.X_min) / (self.X_max - self.X_min)

    def train(self):
        X_b = np.column_stack([np.ones(self.X_train.shape[0]), self.X_train])
        theta, _, _, _ = np.linalg.lstsq(X_b, self.y_train, rcond=None)
        self.intercept_ = theta[0]
        self.coef_ = theta[1:]

    def evaluate(self):
        y_pred = self.predict(self.X_test)
        mse = np.mean((self.y_test - y_pred)**2)
        ss_tot = np.sum((self.y_test - np.mean(self.y_test))**2)
        r2 = 1 - np.sum((self.y_test - y_pred)**2)/ss_tot
        print(f"MSE: {mse:.3f}, R²: {r2:.3f}")

    def predict(self, X):
        return X @ self.coef_ + self.intercept_

    def run(self, X, y, test_ratio=0.2):
        self.load_data(X, y)
        self.train_test_split(test_ratio)
        self.normalize()
        self.train()
        self.evaluate()


np.random.seed(42)
X = np.random.rand(50, 2)*10
y = 3*X[:,0] - 2*X[:,1] + 5 + np.random.randn(50)*0.5

pipeline = MLPipeline()
pipeline.run(X, y)
````
