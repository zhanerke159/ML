**Жағдаят:** Банктің деректері бойынша несие алушылардың 8%-ы несиені қайтармайды (дефолт). Банктің скоринг жүйесі дефолт болған жағдайда 90% дұрыстықпен ескертеді, ал дефолт болмаса 15% жалған ескерту береді.

**Тапсырма:** Байес теоремасы арқылы жүйе ескертсе, шынымен дефолт болу ықтималдығын есептеңіз.

Орындаңыз:

1. Байес теоремасының формуласын кодта іске асырыңыз
2. Жүйе ескертсе, дефолт болу ықтималдығын есептеңіз
3. Дефолт таралуы 1%-дан 30%-ға дейін өзгергенде апостериорлы ықтималдықтың өзгерісін графикте көрсетіңіз
4. Қандай таралу деңгейінде апостериорлы ықтималдық 50%-дан асады?

```python
# 4-кейс деректері (өзгертпеңіз)
p_default = 0.08           # дефолт таралуы: 8%
p_alert_given_default = 0.90    # дефолт болса, жүйе ескертеді: 90%
p_alert_given_no_default = 0.15  # дефолт болмаса, жалған ескерту: 15%

print('Банк несиесі кейсі — берілген мәліметтер:')
print(f'Дефолт таралуы: {p_default:.0%}')
print(f'Дұрыс ескерту ықтималдығы: {p_alert_given_default:.0%}')
print(f'Жалған ескерту ықтималдығы: {p_alert_given_no_default:.0%}')
```

```python
# 4-кейс шешімі
# Осы жерге кодыңызды жазыңыз

# 1. Байес теоремасы
# P(дефолт | ескерту) = P(ескерту | дефолт) * P(дефолт) / P(ескерту)
# TODO:
p_not_def = 1-p_default
p_alert = p_alert_given_default * p_default + p_alert_given_no_default * p_not_def
p_def_given_alert = p_alert_given_default * p_default / p_alert
print("P(D|W): ",p_def_given_alert)

# 2. Таралу диапазонында апостериорлы ықтималдық графигі
# TODO:
pr = np.linspace(0, 1, 100)
prs = []
for p in pr:
  p_not_def = 1-p
  p_alert = p_alert_given_default * p + p_alert_given_no_default * p_not_def
  p_def_given_alert = p_alert_given_default * p / p_alert
  prs.append(p_def_given_alert)

plt.figure(figsize=(8,5))
plt.plot(pr, prs)
plt.axhline(0.5, linestyle='--')
plt.title("Апостериорлы ықтималдықтың өзгеруі")
plt.xlabel("Дефолттың бастапқы ықтималдығы")
plt.ylabel("P(Дефолт | Ескерту)")
plt.grid(True)
plt.show()


# 3. 50% шегін кесіп өту нүктесі
# TODO:
for pd, p in zip(pr, prs):
  if p > 0.5:
    print(pd)
```

**Жағдаят:** Университет деканаты 3 пәннің емтихан нәтижелерін талдауды тапсырды: математика, бағдарламалау, статистика. Деректерді стандартизациялап, пәндер арасындағы байланысты анықтау керек.

**Тапсырма:** Z-score стандартизациясын қолданып, пәндер арасындағы корреляцияны талдаңыз.

Орындаңыз:

1. Берілген деректерге StandardScaler қолданыңыз
2. Стандартизацияланған деректердің орташа мәні мен стандарт ауытқуын тексеріңіз
3. 3 пән арасындағы корреляция матрицасын есептеңіз
4. Корреляция матрицасының жылу картасын (heatmap) сызыңыз
5. Ең жоғары корреляция қай пәндер арасында?

```python
# 5-кейс деректері (өзгертпеңіз)
np.random.seed(15)  # тұрақты деректер үшін тұқым
n_students = 80     # студенттер саны

# Математика бағасы
math_scores = np.random.normal(68, 12, n_students)
# Бағдарламалау: математикамен байланысты
programming_scores = 0.7 * math_scores + np.random.normal(10, 8, n_students)
# Статистика: математикамен орташа байланысты
stats_scores = 0.5 * math_scores + np.random.normal(15, 10, n_students)

# Бағаларды 0-100 диапазонына шектеу
math_scores = np.clip(math_scores, 0, 100)
programming_scores = np.clip(programming_scores, 0, 100)
stats_scores = np.clip(stats_scores, 0, 100)

exam_data = np.column_stack([math_scores, programming_scores, stats_scores])
subject_names = ['Математика', 'Бағдарламалау', 'Статистика']

print(f'Студенттер саны: {n_students}')
print(f'Математика орташасы: {math_scores.mean():.1f}')
print(f'Бағдарламалау орташасы: {programming_scores.mean():.1f}')
print(f'Статистика орташасы: {stats_scores.mean():.1f}')
```

```python
# 5-кейс шешімі
# Осы жерге кодыңызды жазыңыз

# 1. StandardScaler қолдану
# TODO:
scaler = StandardScaler()
scaler_data = scaler.fit_transform(exam_data)

# 2. Стандартизацияны тексеру (орташа ~ 0, std ~ 1)
# TODO:
print(scaler_data.mean(axis=0))
print(scaler_data.std(axis=0))

# 3. Корреляция матрицасы
# TODO:
corr_m = np.corrcoef(scaler_data, rowvar=False)


# 4. Жылу картасы
# TODO:
sns.heatmap(corr_m, annot=True, cmap='coolwarm', xticklabels=subject_names, yticklabels=subject_names)
plt.title("Пәндер арасындағы коррелация")
plt.show()

print('Шешімді осы жерге жазыңыз...')
print("Корреляция матрицасы: ")
print(corr_m)
```

**Жағдаят:** Технологиялық компания 4 смартфонды салыстыру үшін олардың характеристикаларын векторлар түрінде берді. Ұсыныс жүйесі (recommendation system) телефондар арасындағы ұқсастықты анықтау үшін нормаларды қолданады.

**Тапсырма:** Векторлар нормасы арқылы смартфондарды салыстырыңыз.

Орындаңыз:

1. Әр смартфонның L1 және L2 нормасын есептеңіз
2. iPhone мен Samsung арасындағы Евклид қашықтығын табыңыз
3. Барлық жұптар арасындағы қашықтық матрицасын есептеңіз
4. Ең ұқсас екі смартфонды анықтаңыз
5. Нормализацияланған векторлар арасындағы косинустық ұқсастықты есептеңіз

```python
# 7-кейс деректері (өзгертпеңіз)
# Характеристикалар: [баға(мың тг), камера(МП), батарея(мАч), жады(ГБ), CPU жылдамдығы(ГГц)]
smartphones = {
    'iPhone 15':    np.array([450, 48, 3877, 128, 3.46]),
    'Samsung S24':  np.array([380, 50, 4000, 256, 3.39]),
    'Xiaomi 14':    np.array([280, 50, 4610, 256, 3.20]),
    'Google Pixel': np.array([350, 50, 4575, 128, 3.00])
}

phone_names = list(smartphones.keys())
feature_labels = ['Баға', 'Камера', 'Батарея', 'Жады', 'CPU']

print('Смартфон характеристикалары:')
for name, features in smartphones.items():
    print(f'{name}: {features}')
```

```python
# 7-кейс шешімі
# Осы жерге кодыңызды жазыңыз

# 1. L1 және L2 нормалары
# TODO:
for name, vec in smartphones.items():
    l1 = np.linalg.norm(vec, ord=1)
    l2 = np.linalg.norm(vec, ord=2)
    print(name, "L1:", l1, "L2:", l2)

# 2. iPhone мен Samsung арасындағы Евклид қашықтығы
# TODO:
e = np.linalg.norm(smartphones['iPhone 15'] - smartphones['Samsung S24'])

# 3. Қашықтық матрицасы (барлық жұптар)
# TODO:
n = len(phone_names)

dist_matrix = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        v1 = smartphones[phone_names[i]]
        v2 = smartphones[phone_names[j]]
        dist_matrix[i,j] = np.linalg.norm(v1 - v2)

print(dist_matrix)
# 4. Ең ұқсас жұп
# TODO:
min_dist = float('inf')
pair = None

for i in range(n):
    for j in range(i+1,n):
        d = np.linalg.norm(
            smartphones[phone_names[i]] -
            smartphones[phone_names[j]]
        )

        if d < min_dist:
            min_dist = d
            pair = (phone_names[i], phone_names[j])

print("Ең ұқсас смартфондар:", pair)

# 5. Косинустық ұқсастық
# TODO:
def cosine_similarity(a,b):
    return np.dot(a,b) / (np.linalg.norm(a)*np.linalg.norm(b))

for i in range(n):
    for j in range(i+1,n):
        v1 = smartphones[phone_names[i]]
        v2 = smartphones[phone_names[j]]

        sim = cosine_similarity(v1,v2)

        print(phone_names[i], "-", phone_names[j], ":", sim)

print('Шешімді осы жерге жазыңыз...')

```

3. Деректер қалыпты таралымға қаншалықты сәйкес? (гистограмма + PDF)
4. 68-95-99.7 ережесін осы деректерге тексеріңіз
5. Нәтижелерді дәрігерге арналған қысқаша есеп түрінде шығарыңыз

```python
# 8-кейс деректері (өзгертпеңіз)
np.random.seed(33)  # тұрақты деректер үшін тұқым
# Пациенттердің систолалық қан қысымы (мм.сн.бағ.)
blood_pressure = np.concatenate([
    np.random.normal(120, 15, 170),  # қалыпты пациенттер
    np.random.normal(155, 10, 20),   # жоғары қысымды пациенттер
    np.random.normal(80, 8, 10)      # төмен қысымды пациенттер
])
blood_pressure = np.clip(blood_pressure, 60, 200)  # физиологиялық шектеу

NORMAL_MIN = 90   # қалыпты минимум
NORMAL_MAX = 140  # қалыпты максимум

print(f'Пациенттер саны: {len(blood_pressure)}')
print(f'Қан қысымы диапазоны: {blood_pressure.min():.1f} - {blood_pressure.max():.1f} мм.сн.бағ.')
print(f'Қалыпты деңгей: {NORMAL_MIN} - {NORMAL_MAX} мм.сн.бағ.')
```

```python
# 8-кейс шешімі
# Осы жерге кодыңызды жазыңыз

# 1. Негізгі статистика
# TODO:
mean_bp = np.mean(blood_pressure)
std_bp = np.std(blood_pressure)

print(mean_bp)
print(std_bp)

# 2. Қауіп аймағындағы пациенттер саны
# TODO:
risk_patients = np.sum((blood_pressure < NORMAL_MIN) | (blood_pressure > NORMAL_MAX))

percent_risk = risk_patients / len(blood_pressure) * 100

print("Қауіп аймағындағы пациенттер саны:", risk_patients)
print("Пайызы:", percent_risk)
# 3. Гистограмма + қалыпты таралым PDF
# TODO:
plt.figure(figsize=(8,5))

sns.histplot(blood_pressure, bins=20, stat="density")

x = np.linspace(blood_pressure.min(), blood_pressure.max(), 200)
pdf = stats.norm.pdf(x, mean_bp, std_bp)

plt.plot(x, pdf)

plt.title("Қан қысымы таралуы")
plt.xlabel("Қан қысымы")
plt.ylabel("Тығыздық")

plt.show()

# 4. 68-95-99.7 ережесін тексеру
# TODO:
within1 = np.mean((blood_pressure > mean_bp - std_bp) & (blood_pressure < mean_bp + std_bp))
within2 = np.mean((blood_pressure > mean_bp - 2*std_bp) & (blood_pressure < mean_bp + 2*std_bp))
within3 = np.mean((blood_pressure > mean_bp - 3*std_bp) & (blood_pressure < mean_bp + 3*std_bp))

print("1σ ішінде:", within1 * 100)
print("2σ ішінде:", within2 * 100)
print("3σ ішінде:", within3 * 100)

# 5. Дәрігерге арналған есеп
# TODO:
print(f"Орташа қан қысымы: {mean_bp:.1f} мм.сн.бағ.")
print(f"Стандарт ауытқу: {std_bp:.1f}")

print(f"Қауіп аймағындағы пациенттер: {risk_patients} ({percent_risk:.1f}%)")

print("68-95-99.7 ережесі:")
print(f"1σ ішінде: {within1*100:.1f}%")
print(f"2σ ішінде: {within2*100:.1f}%")
print(f"3σ ішінде: {within3*100:.1f}%")
print('Шешімді осы жерге жазыңыз...')
```



# 9-кейс: Тұрғын Үй Нарығы — Меншікті Мәндер

**Жағдаят:** Жылжымайтын мүлік агенттігі үй бағасына әсер ететін 3 фактордың ковариация матрицасын жинады: ауданы, бөлме саны, орталыққа қашықтығы. ML инженері PCA жасамас бұрын матрицаның меншікті мәндерін талдауды сұрады.

**Тапсырма:** Ковариация матрицасының меншікті мәндері мен векторларын талдаңыз.

Орындаңыз:

1. Берілген деректер матрицасынан ковариация матрицасын есептеңіз
2. Ковариация матрицасының меншікті мәндері мен векторларын табыңыз
3. Меншікті мәндерді кему ретімен сорттаңыз
4. Әр меншікті мәннің жалпы дисперсияға үлесін (explained variance ratio) есептеңіз
5. Үлесті графикте (bar chart) көрсетіңіз

```python
# 9-кейс деректері (өзгертпеңіз)
np.random.seed(21)  # тұрақты деректер үшін тұқым
n_houses = 120      # үйлер саны

# Үй характеристикалары
area = np.random.normal(85, 25, n_houses)             # ауданы (м²)
rooms = np.round(area / 25 + np.random.normal(0, 0.5, n_houses))  # бөлме саны
distance = np.random.normal(10, 5, n_houses)          # орталыққа қашықтық (км)

housing_matrix = np.column_stack([area, rooms, distance])
housing_features = ['Аудан (м²)', 'Бөлме саны', 'Қашықтық (км)']

print(f'Үйлер саны: {n_houses}')
print(f'Аудан диапазоны: {area.min():.1f} - {area.max():.1f} м²')
print(f'Бөлме саны диапазоны: {rooms.min():.0f} - {rooms.max():.0f}')
print(f'Қашықтық диапазоны: {distance.min():.1f} - {distance.max():.1f} км')
```

```python
# 9-кейс шешімі
# Осы жерге кодыңызды жазыңыз

# 1. Ковариация матрицасы
# TODO:
cov_matrix = np.cov(housing_matrix, rowvar=False)

print("Ковариация матрицасы:")
print(cov_matrix)
# 2. Меншікті мәндер мен векторлар
# TODO:
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Меншікті мәндер:", eigenvalues)
print("Меншікті векторлар:")
print(eigenvectors)

# 3. Кему ретімен сорттау
# TODO:
idx = np.argsort(eigenvalues)[::-1]

eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# 4. Explained variance ratio
# TODO:
e= eigenvalues / np.sum(eigenvalues)

print(e)

# 5. Bar chart
# TODO:
plt.figure(figsize=(6,4))

plt.bar(range(1,4), e)

plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Variance explained by components")

plt.show()

print('Шешімді осы жерге жазыңыз...')
```



# 10-кейс: Жиынтық Кейс — Электр Энергиясын Тұтыну Талдауы

**Жағдаят:** Энергетика компаниясы 5 аудандағы 30 күндік электр энергиясын тұтыну деректерін (кВт/сағат) жинады. Компания осы деректерді ML моделіне беру алдында толық алдын ала өңдеу жасағысы келеді.

**Тапсырма:** Барлық дәрісте өткен тақырыптарды бір жобада біріктіріп қолданыңыз.

Орындаңыз:

1. Деректер матрицасын жасаңыз (30 күн x 5 аудан)
2. Әр аудан үшін сипаттамалық статистика есептеңіз (орташа, медиана, std, min, max)
3. Min-Max және Z-score нормализацияларын екеуін де қолданыңыз
4. Аудандар арасындағы корреляция матрицасын есептеп, жылу картасын сызыңыз
5. Тұтыну деректерінің таралымын гистограмма арқылы визуализациялаңыз (5 аудан = 5 subplot)
6. Ең жоғары және ең төмен тұтыну күндерін анықтаңыз
7. Нәтижелерді жиынтық кесте түрінде шығарыңыз

```python
# 10-кейс деректері (өзгертпеңіз)
np.random.seed(99)  # тұрақты деректер үшін тұқым
n_days = 30         # күндер саны

# Аудандардың тұтыну профильдері (кВт/сағат)
district_data = {
    'Алатау':    np.random.normal(450, 60, n_days),   # тұрғын үй ауданы
    'Бостандық': np.random.normal(380, 45, n_days),   # аралас аудан
    'Медеу':     np.random.normal(520, 80, n_days),   # өнеркәсіп ауданы
    'Наурызбай': np.random.normal(290, 35, n_days),   # жаңа тұрғын аудан
    'Түрксіб':   np.random.normal(610, 95, n_days)    # ескі өнеркәсіп ауданы
}

# Деректер матрицасы: 30 күн x 5 аудан
energy_matrix = np.column_stack(list(district_data.values()))
district_names = list(district_data.keys())

print(f'Деректер матрицасы өлшемі: {energy_matrix.shape}')
print(f'Аудандар: {district_names}')
print(f'\nБастапқы деректер үзіндісі (алғашқы 5 күн):')
print(energy_matrix[:5, :])
```

```python
# 10-кейс шешімі — 1-бөлім: Сипаттамалық Статистика
# Осы жерге кодыңызды жазыңыз

# Әр аудан үшін: орташа, медиана, std, min, max
# TODO:

print('Шешімді осы жерге жазыңыз...')
stats_table = []

for i, name in enumerate(district_names):

    data = energy_matrix[:, i]

    mean = np.mean(data)
    median = np.median(data)
    std = np.std(data)
    min_val = np.min(data)
    max_val = np.max(data)

    stats_table.append([name, mean, median, std, min_val, max_val])
for row in stats_table:
    print(row)
```

```python
# 10-кейс шешімі — 2-бөлім: Нормализация
# Осы жерге кодыңызды жазыңыз

# Min-Max нормализациясы
# TODO:
minmax = MinMaxScaler()

energy_minmax = minmax.fit_transform(energy_matrix)

# Z-score стандартизациясы
# TODO:
scaler = StandardScaler()

energy_zscore = scaler.fit_transform(energy_matrix)

print('Шешімді осы жерге жазыңыз...')
print(energy_minmax)
print(energy_zscore)
```

```python
# 10-кейс шешімі — 3-бөлім: Корреляция және Визуализация
# Осы жерге кодыңызды жазыңыз

# Корреляция матрицасы + жылу картасы
# TODO:
corr_matrix = np.corrcoef(energy_matrix, rowvar=False)
print(corr_matrix)

# 5 аудан гистограммасы (subplot)
# TODO:
plt.figure(figsize=(12,8))

for i, name in enumerate(district_names):

    plt.subplot(2,3,i+1)

    plt.hist(energy_matrix[:, i], bins=10)

    plt.title(name)
    plt.xlabel("Тұтыну")
    plt.ylabel("Күн саны")

plt.tight_layout()
plt.show()
# Ең жоғары/төмен тұтыну күндері
# TODO:
max_day = np.argmax(energy_matrix)
min_day = np.argmin(energy_matrix)

max_value = energy_matrix.flatten()[max_day]
min_value = energy_matrix.flatten()[min_day]

print("Ең жоғары тұтыну:", max_value)
print("Ең төмен тұтыну:", min_value)
# Жиынтық кесте
# TODO:
for i, name in enumerate(district_names):

    data = energy_matrix[:, i]

    mean = np.mean(data)
    median = np.median(data)
    std = np.std(data)
    min_val = np.min(data)
    max_val = np.max(data)

    print(f"{name} | {mean:.2f} | {median:.2f} | {std:.2f} | {min_val:.2f} | {max_val:.2f}")
print('Шешімді осы жерге жазыңыз...')
```
