**1-тапсырма (ReLU активация функциясы):**
`relu(x)` функциясын жазыңыз. Егер кіріс мәні `x` нөлден үлкен болса, функция сол санның өзін қайтаруы керек. Егер нөлге тең немесе одан кіші болса, `0` мәнін қайтаруы тиіс.


```
# Осында кодыңызды жазыңыз
def relu(x):
    if x > 0:
        return x
    else:
        return 0
```



**2-тапсырма (Сигмоид активация функциясы):**
`math` модулін импорттаңыз. Логистикалық регрессияда қолданылатын `sigmoid(x)` функциясын жазыңыз. Формуласы: . Мұндағы  үшін `math.exp()` функциясын қолданыңыз.



```
import math
# Осында кодыңызды жазыңыз
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
print(sigmoid(0))
print(sigmoid(2))
```


**3-тапсырма (Бинарлы кросс-энтропия / Binary Cross-Entropy Loss):**
Бір ғана бақылау үшін қателікті есептейтін `calculate_bce(y_true, y_pred)` функциясын жазыңыз. Мұндағы `y_true` нақты класс (0 немесе 1), ал `y_pred` модельдің болжамды ықтималдығы. Формула: . Логарифм үшін `math.log()` пайдаланыңыз.



```
import math
# Осында кодыңызды жазыңыз

def calculate_bce(y_true, y_pred):
    return - (y_true * math.log(y_pred) + (1 - y_true) * math.log(1 - y_pred))
print(calculate_bce(1, 0.9))
print(calculate_bce(0, 0.2))
```



**4-тапсырма (Градиенттік түсудің бірнеше қадамы):**
Бастапқы салмақ `weight = 0.8`, оқыту жылдамдығы `learning_rate = 0.01` және тұрақты градиент `gradient = 1.5` берілген. `for` циклін пайдаланып, салмақты 10 рет жаңартыңыз. Формула: . Әр қадам сайын жаңа салмақты экранға шығарып отырыңыз.



```
# Осында кодыңызды жазыңыз
weight = 0.8
learning_rate = 0.01
gradient = 1.5

for step in range(1, 11):
    weight = weight - learning_rate * gradient
    print(f"Қадам {step}: салмақ = {weight}")
```

**5-тапсырма (Ерте тоқтату / Early Stopping симуляциясы):**
Бастапқы қателік `loss = 5.0` мәніне тең. `while` циклін пайдаланып, әр итерация сайын `loss` мәнін `0.4`-ке азайтыңыз. Егер `loss` мәні `1.0`-ден кіші немесе тең болса, циклді тоқтатып, "Модель оқытылды!" деп шығарыңыз.


```
# Осында кодыңызды жазыңыз
loss = 5.0

while True:
    loss -= 0.4
    print("loss =", loss)

    if loss <= 1.0:
        print("Модель оқытылды!")
        break
```

**6-тапсырма (Оқыту жылдамдығын төмендету / Learning Rate Decay):**
`decay_lr(initial_lr, decay_rate, epoch)` функциясын жазыңыз. Бұл модель ұзақ оқытылған сайын оның қадамдарын кішірейтетін әдіс. Формула: . `epoch` аргументіне әдепкі мән (default value) ретінде `1` беріңіз.


```
# Осында кодыңызды жазыңыз
def decay_lr(initial_lr, decay_rate, epoch=1):
    return initial_lr / (1 + decay_rate * epoch)
  print(decay_lr(0.01, 0.1))
print(decay_lr(0.01, 0.1, 5))
```

**7-тапсырма (Бинарлы жіктеу шегі / Classification Threshold):**
`predict_class(probability, threshold)` функциясын жазыңыз. Егер `probability` (ықтималдық) мәні `threshold` (шек) мәнінен үлкен немесе тең болса `1` санын, әйтпесе `0` санын қайтарыңыз. Мұнда `threshold` үшін әдепкі мәнді `0.5` етіп орнатыңыз.


```
# Осында кодыңызды жазыңыз

def predict_class(probability, threshold=0.5):
    if probability >= threshold:
        return 1
    else:
        return 0
print(predict_class(0.8))
print(predict_class(0.3))
print(predict_class(0.45, 0.4))
```


**8-тапсырма (Абсолютті қателікті шартпен есептеу):**
Нақты мән `y_true = 10` және болжам `y_pred = 15`. Дайын `abs()` функциясын қолданбай, `if-else` шарты арқылы осы екі санның абсолютті айырмашылығын (әрдайым оң сан шығуы керек) есептейтін логика жазыңыз және нәтижесін шығарыңыз.


```
# Осында кодыңызды жазыңыз
y_true = 10
y_pred = 15

if y_true >= y_pred:
    error = y_true - y_pred
else:
    error = y_pred - y_true

print(error)
```


**9-тапсырма (L2 Регуляризация айыппұлы / Penalty):**
Екі жеке салмақ айнымалысы `w1` және `w2`, сондай-ақ `lambda_param` (регуляризация коэффициенті) берілген. Осы үш аргументті қабылдап, L2 регуляризациясының айыппұлын есептейтін функция жазыңыз. Формула: .


```
# Осында кодыңызды жазыңыз
def l2_penalty(w1, w2, lambda_param):
    return lambda_param * (w1**2 + w2**2)
print(l2_penalty(0.8, -1.2, 0.1))
```


**10-тапсырма (Қос цикл / Бұтақталған шешімдер ағашының қарапайым моделі):**
`x1` және `x2` деген екі кіріс айнымалысы бар (мысалы, `x1 = 5`, `x2 = -2`). `if-elif-else` шарттарын пайдаланып кірістірілген (nested) логика жазыңыз:

* Егер `x1 > 0` болса:
* ішінде тағы тексеріңіз, егер `x2 > 0` болса, "Класс A" деп шығарыңыз.
* әйтпесе "Класс B" деп шығарыңыз.


* Егер `x1 <= 0` болса:
* "Класс C" деп шығарыңыз.



```
# Осында кодыңызды жазыңыз
x1 = 5
x2 = -2

if x1 > 0:
    if x2 > 0:
        print("Класс A")
    else:
        print("Класс B")
else:
    print("Класс C")
```
