
**19-тапсырма:** Оптимизатордағы импульсті жаңарту. Үш аргумент (`velocity`, `gradient`, `momentum`) қабылдап, `(momentum * velocity) + gradient` формуласымен жаңа жылдамдықты қайтаратын функция жазыңыз.


```
# Осында кодыңызды жазыңыз
def update_velocity(velocity, gradient, momentum):
    return (momentum * velocity) + gradient

print(update_velocity(0.5, 0.1, 0.9))
```

**20-тапсырма:** Leaky ReLU активациясы. Екі аргумент: `x` және әдепкі мәні `0.01` болатын `alpha` қабылдайтын `leaky_relu` функциясын жасаңыз. Егер `x > 0` болса `x`-ті, әйтпесе `x * alpha` мәнін қайтарыңыз.


```
# Осында кодыңызды жазыңыз
def leaky_relu(x, alpha=0.01):
    if x > 0:
        return x
    else:
        return x * alpha

print(leaky_relu(-10))
print(leaky_relu(10))
```


**21-тапсырма:** Дәлдікті (Accuracy) есептеу. `correct_predictions` және `total_predictions` аргументтерін қабылдап, олардың қатынасын қайтаратын функция жазыңыз. Егер `total_predictions` нөлге тең болса, `0` қайтаратын шарт қосыңыз.

```
# Осында кодыңызды жазыңыз
def accuracy(correct_predictions, total_predictions):
    if total_predictions == 0:
        return 0
    return correct_predictions / total_predictions

print(accuracy(45, 50))
```

**22-тапсырма:** Асыра сілтеуді (Overfitting) тексеру. `train_loss` және `val_loss` қабылдайтын логикалық функция жазыңыз. Егер валидациялық қателік оқыту қателігінен үлкен болса `True`, әйтпесе `False` қайтарыңыз.


```
# Осында кодыңызды жазыңыз
def is_overfitting(train_loss, val_loss):
    return val_loss > train_loss

print(is_overfitting(0.2, 0.3))
```


**23-тапсырма:** Деректерді стандарттау (Z-score). `value`, `mean` (орташа мән) және `std` (стандартты ауытқу) аргументтерін қабылдап, `(value - mean) / std` формуласы бойынша жаңа мәнді қайтаратын `standardize_feature` функциясын құрыңыз.


```
# Осында кодыңызды жазыңыз
def standardize_feature(value, mean, std):
    return (value - mean) / std

print(standardize_feature(10, 5, 2))
```


**24-тапсырма:** Оқыту фазасын анықтау. `epoch` (дәуір) аргументін қабылдап, егер ол 5-тен кіші болса "Warmup" (Қыздыру) мәтінін, әйтпесе "Fine-tuning" (Нақты баптау) мәтінін қайтаратын функция жазыңыз.


```
# Осында кодыңызды жазыңыз
def training_phase(epoch):
    if epoch < 5:
        return "Warmup"
    else:
        return "Fine-tuning"

print(training_phase(3))
print(training_phase(10))
```


**25-тапсырма:** Huber Loss қателігінің қарапайым нұсқасы. `error` және `delta` аргументтерін қабылдайтын функция жазыңыз. Егер қателіктің абсолюттік мәні (`abs(error)`) `delta`-дан кіші немесе тең болса, `0.5 * error ** 2` қайтарыңыз. Әйтпесе, `delta * abs(error) - 0.5 * delta ** 2` формуласын қайтарыңыз.


```
# Осында кодыңызды жазыңыз
def huber_loss(error, delta):
    if abs(error) <= delta:
        return 0.5 * (error ** 2)
    else:
        return delta * abs(error) - 0.5 * (delta ** 2)

print(huber_loss(0.5, 1))
print(huber_loss(3, 1))
```
