## І деңгей: Жеңіл тапсырмалар

Бұл тапсырмалар негізгі синтаксис пен қарапайым шарттарды тексеруге арналған.

```
# 1-тапсырма: Екі санның қосындысы.
# a және b сандарының қосындысын шығарыңыз.
a = 15
b = 27
# Кодты осы жерге жазыңыз:
print(a+b)
```

```
# 2-тапсырма: Жұп немесе Тақ (Even or Odd).
# num саны жұп болса "Even", тақ болса "Odd" деп шығарыңыз.
num = 42
# Кодты осы жерге жазыңыз:
if(num%2==0):
  print("Even")
else:
  print("Odd")
```

```
# 3-тапсырма: Санның таңбасы.
# value саны оң болса "Positive", теріс болса "Negative", нөл болса "Zero" деп шығарыңыз.
value = -8
# Кодты осы жерге жазыңыз:
if(value < 0):
  print("Negative")
elif(value >0):
  print("Positive")
else:
  print("Zero")
```

```
# 4-тапсырма: Деректер типін түрлендіру.
# text_num айнымалысын бүтін санға (int) айналдырып, оған 10-ды қосып шығарыңыз.
text_num = "150"
# Кодты осы жерге жазыңыз:
print(int(text_num)+10)
```

```
# 5-тапсырма: Судың қайнауы.
# temp температурасы 100 немесе одан жоғары болса True, әйтпесе False деп шығарыңыз (bool типінде).
temp = 105
# Кодты осы жерге жазыңыз:
if(temp>100):
  print(True)
else
  print(False)
```

```
# 6-тапсырма: Ұзындықтарды салыстыру.
# word1 және word2 сөздерінің ұзындығы тең болса "Equal", тең болмаса "Not Equal" деп шығарыңыз.
word1 = "Python"
word2 = "Coding"
# Кодты осы жерге жазыңыз:
if(word1 == word2):
  print("Equal")
else:
  print("Not Equal")
```

```
# 7-тапсырма: Бас әріптер.
# sentence мәтінін толығымен бас әріптермен шығарыңыз.
sentence = "hello world"
# Кодты осы жерге жазыңыз:
print(sentence.upper())
```

```
# 8-тапсырма: Мәтінді кесу (Slicing).
# word сөзінің алғашқы 3 әрпін шығарыңыз.
word = "Algorithm"
# Кодты осы жерге жазыңыз:
print(word[:3])
```

```
# 9-тапсырма: Логикалық оператор (AND).
# age 18-ден үлкен немесе тең ЖӘНЕ has_id True болса "Access Granted", әйтпесе "Access Denied" деп шығарыңыз.
age = 20
has_id = True
# Кодты осы жерге жазыңыз:
if(age>=18 and has_id == True):
  print("Access Granted")
else:
  print("Access Denied")
```

```
# 10-тапсырма: Мәтінді біріктіру.
# first және last айнымалыларын ортасына бос орын қосып біріктіріп шығарыңыз.
first = "Steve"
last = "Jobs"
# Кодты осы жерге жазыңыз:
print(first + " " + last)
```

## ІІ деңгей: Орташа тапсырмалар

Бұл бөлімде күрделірек шарттар, кірістірілген (nested) `if/else` және мәтіндермен тереңірек жұмыс істеу талап етіледі.


```
# 11-тапсырма: Үш санның ең үлкені (Max of three).
# x, y, z сандарының ішінен ең үлкенін тауып шығарыңыз (max() функциясын қолданбай, if-else арқылы).
x = 14
y = 28
z = 21
# Кодты осы жерге жазыңыз:
if(x>y and x>z):
  print(x)
elif(y>x and y>z):
  print(y)
else:
  print(z)
```

```
# 12-тапсырма: Кібісе жыл (Leap Year).
# year кібісе жыл екенін тексеріңіз. Шарт: 400-ге бөлінсе немесе (4-ке бөлініп, 100-ге бөлінбесе) - "Leap", әйтпесе "Not Leap".
year = 2024
# Кодты осы жерге жазыңыз:
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap")
else:
    print("Not Leap")
```

```
# 13-тапсырма: Палиндром тексеру.
# s мәтіні оң жағынан да, сол жағынан да бірдей оқылатынын тексеріңіз. Иә болса True, жоқ болса False шығарыңыз.
s = "radar"
# Кодты осы жерге жазыңыз:

print(s == s[::-1])
```

```
# 14-тапсырма: Дауысты немесе дауыссыз әріп.
# char әрпі дауысты (a, e, i, o, u) болса "Vowel", әйтпесе "Consonant" деп шығарыңыз. Әріпті кіші регистрге айналдырып барып тексеріңіз.
char = "E"
# Кодты осы жерге жазыңыз:

if char in "aeiou":
    print("Vowel")
else:
    print("Consonant")
```

```
# 15-тапсырма: Баға қою жүйесі (Grades).
# score бойынша баға шығарыңыз: 90-100 -> "A", 80-89 -> "B", 70-79 -> "C", 60-69 -> "D", 60-тан төмен -> "F".
score = 85
# Кодты осы жерге жазыңыз:

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

```
# 16-тапсырма: Үшбұрыштың жарамдылығы.
# Үшбұрыштың a, b, c қабырғалары берілген. Кез келген екі қабырғасының қосындысы үшіншісінен үлкен болса "Valid", әйтпесе "Invalid" шығарыңыз.
a_side = 5
b_side = 7
c_side = 10
# Кодты осы жерге жазыңыз:
if a_side + b_side > c_side and a_side + c_side > b_side and b_side + c_side > a_side:
    print("Valid")
else:
    print("Invalid")
```

```
# 17-тапсырма: Билет бағасы.
# Егер адам жасы (customer_age) 12-ден кіші болса билет 5$, 12 мен 60 аралығында болса 10$, 60-тан үлкен болса 7$. Бағаны шығарыңыз.
customer_age = 65
# Кодты осы жерге жазыңыз:
if customer_age < 12:
    print(5)
elif customer_age <= 60:
    print(10)
else:
    print(7)
```

```
# 18-тапсырма: Сөздің құрылымы.
# string_val мәтіні "Py" деп басталып, "on" деп аяқталса "Match", әйтпесе "No Match" шығарыңыз.
string_val = "Python"
# Кодты осы жерге жазыңыз:
if string_val.startswith("Py") and string_val.endswith("on"):
    print("Match")
else:
    print("No Match")
```

```
# 19-тапсырма: Бөлшек санды дөңгелектеусіз кесу.
# float_val санының тек бүтін бөлігі мен бөлшек бөлігін бөліп шығарыңыз (мысалы: "Бүтін: 7, Бөлшек: 0.85"). int() және арифметиканы қолданыңыз.
float_val = 7.85
# Кодты осы жерге жазыңыз:
integer_part = int(float_val)
fractional_part = float_val - integer_part

print("Бүтін:", integer_part)
print("Бөлшек:", fractional_part)
```

```
# 20-тапсырма: Айнымалыларды ауыстыру (Swap).
# var1 және var2 мәндерін үшінші айнымалыны қолданбай ауыстырыңыз және нәтижесін шығарыңыз.
var1 = 100
var2 = 200
var1, var2 = var2, var1

print("var1 =", var1)
print("var2 =", var2)
```

## ІІІ деңгей: Күрделі тапсырмалар

Бұл тапсырмалар алгоритмдік ойлауды, математикалық логиканы және мәтінмен күрделірек жұмыс жасауды қажет етеді.


```
# 21-тапсырма: FizzBuzz логикасы.
# n саны 3-ке қалдықсыз бөлінсе "Fizz", 5-ке бөлінсе "Buzz", екеуіне де бөлінсе "FizzBuzz", ешқайсысына бөлінбесе санның өзін мәтін түрінде шығарыңыз.
n = 15
# Кодты осы жерге жазыңыз:
if n % 3 == 0 and n % 5 == 0:
    print("FizzBuzz")
elif n % 3 == 0:
    print("Fizz")
elif n % 5 == 0:
    print("Buzz")
else:
    print(str(n))
```

```
# 22-тапсырма: Қарапайым калькулятор.
# num1, num2 сандары және operator ('+', '-', '*', '/') берілген. Шарттарды пайдаланып дұрыс математикалық амалды орындап, нәтижесін шығарыңыз. (Нөлге бөлу қателігін де ескеріңіз: егер '/' болса және num2 == 0 болса, "Error" шығарыңыз).
num1 = 20
num2 = 4
operator = "/"
# Кодты осы жерге жазыңыз:
if operator == "+":
    print(num1 + num2)
elif operator == "-":
    print(num1 - num2)
elif operator == "*":
    print(num1 * num2)
elif operator == "/":
    if num2 == 0:
        print("Error")
    else:
        print(num1 / num2)
else:
    print("Error")
```

```
# 23-тапсырма: Уақытты түрлендіру.
# total_seconds берілген. Осыны "Сағат:Минут:Секунд" (HH:MM:SS) форматына түрлендіріп шығарыңыз. (Бүтін бөлу // және қалдық % операторларын қолданыңыз).
total_seconds = 3665
# Кодты осы жерге жазыңыз:

hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60

print(f"{hours:02}:{minutes:02}:{seconds:02}")
```

```
# 24-тапсырма: Құпия сөздің мықтылығы.
# pwd мәтіні берілген. Егер оның ұзындығы 8-ден үлкен болса ЖӘНЕ ішінде "!" немесе "@" белгілерінің бірі болса "Strong", әйтпесе "Weak" деп шығарыңыз.
pwd = "MyPassword@1"
# Кодты осы жерге жазыңыз:
if len(pwd) > 8 and ("!" in pwd or "@" in pwd):
    print("Strong")
else:
    print("Weak")
```

```
# 25-тапсырма: Ортаңғы әріп(тер)ді табу.
# text мәтінінің ұзындығы тақ болса қақ ортасындағы 1 әріпті, жұп болса ортасындағы 2 әріпті шығарыңыз. (Slicing және // операторларын қолданыңыз).
text = "Django" # Ұзындығы 6 (жұп), демек "an" шығуы керек. ("Python" -> "th", "Apple" -> "p")
# Кодты осы жерге жазыңыз:
n = len(text)
m = n // 2

if n % 2 == 1:
    print(text[m])
else:
    print(text[m-1:m+1])
```
